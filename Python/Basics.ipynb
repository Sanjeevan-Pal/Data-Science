# 3. Variables

```python
# Static Vs Dynamic Typing
# Static Vs Dynamic Binding
# stylish declaration techniques
```

```python
# C/C++
name = 'nitish'
print(name)

a = 5
b = 6

print(a + b)
```

```python
# Dynamic Typing, we dont have to specify the name of the datatype
a = 5
# Static Typing, we have to specify the name of the datatype(happens in C, Java)
#int a = 5
```

```python
# Dynamic Binding, variable's datatype can be changed
a = 5
print(a)
a = 'nitish'
print(a)

# Static Binding, the variable's can not be changed
#int a = 5
```

```python
#Multiple printing and declaring of variable
a = 1
b = 2
c = 3
print(a,b,c)
```

```python
a,b,c = 1,2,3
print(a,b,c)
```

```python
#Assigning various variables same value
a=b=c= 5
print(a,b,c)
```

# 7. Literals

```python
a = 0b1010 #Binary Literals
b = 100 #Decimal Literal
c = 0o310 #Octal Literal
d = 0x12c #Hexadecimal Literal

#Float Literal
float_1 = 10.5
float_2 = 1.5e2 # 1.5 * 10^2
float_3 = 1.5e-3 # 1.5 * 10^-3

#Complex Literal
x = 3.14j

print(a, b, c, d)
print(float_1, float_2,float_3)
print(x, x.imag, x.real)
```

```python
# binary
x = 3.14j
print(x.imag)
```
