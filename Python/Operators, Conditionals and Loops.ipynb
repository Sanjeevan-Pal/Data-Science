# Operators in Python

- Arithmetic Operators
- Relational Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Membership Operators

```python
# Arithmetric Operators
print(5+6)

print(5-6)

print(5*6)

print(5/2)

print(5//2)

print(5%2)

print(5**2)
```

```python
# Relational Operators
#Returns True or False
print(4>5)

print(4<5)

print(4>=4)

print(4<=4)

print(4==4)

print(4!=4)
```

```python
# Logical Operators
#True=1
#False=0
print(1 and 0)

print(1 or 0)

print(not 1)
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

print(~3)

print(4 >> 2)

print(5 << 2)
```

```python
# Assignment Operators

# =
# a = 2

a = 2

# a = a % 2
a %= 2

# a++ ++a

print(a)
```

```python
# Membership Operators

# in/not in

print('D' not in 'Delhi')

print(1 in [2,3,4,5,6])
```

```python
# Program - Find the sum of a 3 digit number entered by the user

number = int(input('Enter a 3 digit number'))
sum=0
while number/10>0:
  sum+=number%10
  number//=10
print(sum)
```

```python
# Guessing game

# generate a random integer between 1 and 100
import random
jackpot=random.randint(1,100)
count=0
guess=int(input("Enter your guess:"))
while guess!=jackpot:
  if guess<jackpot:
    print("You guessed  lower number. Guess a higher number.")
  else:
    print("You guessed higher number. Guess a lower number.")
  count+=1
  guess=int(input("Enter your guess:"))
print("Correct guess")
print("Number of attempts is",count)
```

### Sequence sum

1/1! + 2/2! + 3/3! + ...

```python
# code here
import math

def sequence_sum(n_terms):
    total = 0
    for i in range(1, n_terms + 1):
        total += i / math.factorial(i)
    return total

# Test it
n = int(input("Enter number of terms: "))
result = sequence_sum(n)
print(f"Sum of {n} terms = {result}")


## Let's Simplify the Math First

```

### Pattern 2

1<br>
121<br>
12321<br>
1234321<br>

```python
n = int(input("Enter number of patterns: "))

for i in range(1, n + 1):
    # Count up: 1 to i
    count_up = ""
    for j in range(1, i + 1):
        count_up += str(j)

    # Count down: i-1 to 1
    count_down = ""
    for j in range(i - 1, 0, -1):
        count_down += str(j)

    print(count_up + count_down, end="\n")
```
